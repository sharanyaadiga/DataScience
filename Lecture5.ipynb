**Statistical population** = a set of objects $S$

**Probability**(measure) $P:2^S\to [0,1]$, such that:

1. $P(\emptyset)=0$
1. $P(S)=1$
1. $P(A\cup B)=P(A)+P(B)$, $\forall A,B\subset S$ with $A\cap B=\emptyset$

(The axioms of Kolmogorov)

where $2^S=\{A \mid A\subset S\}$ is the set of subsets of $S$.

* Note: P could be partial, not every subset of $S$ measurable

**(statistical) random variable** = a function $X:S\to V$

The objects have (named) attributes/features

**Obs.** Each object attribute is a random variable as well

Types of variables:

- **qualitative (categorical)**: e.g. good/bad ; red/yellow/blue ; young/old ; sometimes coded with numbers
        represents a quality
        
- **quantitative**: e.g. -20 Celsius ; 100 m ; 100 Euro
        represents an amount
        
By the set of values:

- **discrete**: a finite set of values or an enumerable set
- **continuous**: takes all values in an interval
        



By existance of the order/distance between values:
- **metric** variables: order and distance exist
    
    e.g. temperature
    
    
- **ordinal** variables: there is an order, but no meaningful distance:
    
    e.g. school grades in Germany: 1,2,3,4,5,6 (1 is better than 2, 2 is better than 3, etc)


- **nominal** variables: the values are names/categories, no order nor distance.

**Statistical event** = Subset $A\subset S$
- Corresponds to a binary statistical variable, $1_A:S\to\{0,1\}$, the indicator function of $A$

    $1_A(x)=1$ if and only if $x\in A$, otherwise $1_A(x)=0$.
   
Obs.: the probability function measures the events. $P(A)$ is the **probability of the event A**.
    

**Independent events**

$A,B\subset S$ are independent when $P(A\cap B)=P(A)*P(B)$

**Induced probability space**

Let $B\subset S$, with $P(B)>0$. An induced probability function on $B$ is the function $P(\cdot\mid B):2^B\to [0,1]$ defined by:
$$P(A\mid B)=\frac{P(A\cap B)}{P(B)}$$

$P(A\mid B)$ is named the **conditional probability** of $A$ given $B$.


Note that for independent events $A, B$ with $P(B)>0$:
$$P(A\mid B)=P(A)$$

**Statistical event derived from a variable which is not binary**
Let $X:S\to B$ be a random variable and $b\in B$ one of its possible values. 
The event "X=b" is defined by this set $\{a\in S\mid X(a)=b\}$

Similarly one defines other events derived from a random variable, 
e.g. $X<b$, $X^2>b$ when $B$ is a set of numbers, always with the meaning of being the subset of $S$ formed by the elements which, when mapped through $X$ satisfy the given predicate.


# Bayes' Theorem

Let $A, B\subset S$ be any two events, with $P(A)>0$. Then

$$P(B\mid A)=\frac{P(A\mid B)*P(B)}{P(A)}$$

Example:

$$P(Disease\mid Symptom)=\frac{P(Symptom\mid Disease)*P(Disease)}{P(Symptom)}$$

**Apriori probability**: $P(Disease)$, over the entire population, before knowing of the symptoms.

**Aposteriori probability**: $P(Disease\mid Symptom)$ the probability of the disease conditioned by having the symptom.

Informally, a way to "invert" conditional probabilities.

## Relation with test Sensitivity, Specificity, Precision, Recall, etc.

```python
tp=117
tn=743
fp=3
fn=2
testpos=120
testneg=745
realpos=119
realneg=746
total=865
print(f"precision={100* tp/testpos}%")
print(f"sensitivity=recall={100* tp/realpos}%  ✓")
print(f"accuracy={100* (tp+tn)/total}% ✓")
print(f"specificity={100* tn/realneg}% ✓")
```

### What is the probability of the sickness, when the test reports "positive" ?

Applying the formla of Bayes:

$$P(realpos \mid testpos)=\frac{P(testpos \mid realpos)*P(realpos)}{P(testpos)}$$

```python
recall=tp/realpos
P_realpos_given_testpos=(recall*(realpos/total))/(testpos/total)
print(f"P_realpos_given_testpos={P_realpos_given_testpos}")
```

The **PredictivePositiveValue** (PPV) -- the quantity we just computed -- is always equal with the precision

## When our intuition fails us

A frequent logical confusion https://en.wikipedia.org/wiki/Confusion_of_the_inverse



[Example from the cited wikipedia page] Assume: 
- cancer prevalence is 1% of the population 
- we have a test that is 80% sensitive (20% FN) and 90% specific (10% FP)
- and a patient with positive test result

What is the probability that the patient has cancer?


What is being asked is the positive predictive value = precision

$$precision=\frac{TP}{TestPos}=P(RealPos \mid TestPos)=?$$

What we do know:

$$recall=sensitivity=\frac{TP}{RealPos}=80\%$$

$$specificity=\frac{TN}{RealNeg}=90\%$$

**and**

$$prevalence=\frac{RealPos}{RealPos+RealNeg}=P(RealPos)=1\%$$

Bayes formula

$$P(RealPos \mid TestPos)=\frac{P(TestPos \mid RealPos)*P(RealPos)}{P(TestPos)}$$

$$P(TestPos)=P(TestPos \mid RealPos)*P(RealPos)+P(TestPos \mid RealNeg)*P(RealNeg)$$

$$P(TestPos)=0.8*0.01+(1-0.9)*0.99=0.10699999999999998$$

$$P(RealPos \mid TestPos)=\frac{0.8*0.01}{0.10699999999999998}=0.07476635514018692 \approx 7.5\%$$

# The Bayes Classifier

**Problem statement**

Let $C$ (from "class") be a random variable with the possible values c1,c2,c3 (two or more, here 3).

Let $F$ be a feature and "a" a possible value.

Our observation of the feature $F$ shown the value "a".

What is the probability of each of the three classes?

Simplified notation:
    
$P(c1):=P(C=c1)$
    
$P(a):=P(F=a)$

$P(c1\mid a)=\frac{P(a\mid c1)*P(c1)}{P(a)} \sim P(a\mid c1)*P(c1)$

$P(c2\mid a)=\frac{P(a\mid c2)*P(c2)}{P(a)} \sim P(a\mid c2)*P(c2)$

$P(c3\mid a)=\frac{P(a\mid c3)*P(c3)}{P(a)} \sim P(a\mid c3)*P(c3)$

Decide for the class with the highest aposteriori probability, use the precise values for computing confidence.

What we eliminated, $P(a)$ is the normalization factor such that $(P(c1\mid a),P(c2\mid a),P(c3\mid a)$ is a discrete probability distribution over the three classes c1, c2 and c3.

$$P(a)= P(a\mid c1)*P(c1) +  P(a\mid c2)*P(c2) +  P(a\mid c3)*P(c3)$$

#### Bayes Classifier, Rule of thumb

$Aposteriori=ConditionalLikelihood * Apriori$

### Median

let $X:S\to \mathbb{R}$ be a random variable with real values.

Then any value $m$ such that:
$$P(X\le m)=P(X\ge m)=0.5$$
is named a median value for X.

### Quantile
let $X:S\to \mathbb{R}$ be a random variable with real values and $p\in [0,1]$.

Then any value $m$ such that:
$$P(X\le m)=p$$
is named a p-level quantile for X.

Obs.: The median corresponds to a 0.5-quantile.
    
We can also find for a given integer $q$ the $q-1$ $q$-quantiles, 
which are the quantiles of level $1/q$, $2/q$, ...$(q-1)/q$.

For $q=4$ quartiles (levels 0.25, 0.5, 0.75).
